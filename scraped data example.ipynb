```python
import snscrape.modules.twitter as snt
import pandas as pd

# Getting Tweets from Ratan TATA

Model_1 = "Twitter"
tweets = []
limit = 500

for tweet in snt.TwitterSearchScraper(Model_1).get_items():
    #print(vars(tweet))
    #break
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date,tweet.id,tweet.url,tweet.user.username,tweet.media,tweet.replyCount,tweet.retweetCount])
            
            
df = pd.DataFrame(tweets, columns = ["DateTime", "ID", "Text","User_Name","Media","ReplyCount","Retweet Count"])
display(df)


    
```
